```python
import math

data = [(1,2), (2,2), (2,3), (8,8), (8,9), (25,25)]

eps = 2
minPts = 2

visited = set()
clusters = []
noise = []

def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

def region_query(point):
    neighbors = []
    for p in data:
        if distance(point, p) <= eps:
            neighbors.append(p)
    return neighbors

def expand_cluster(point, neighbors, cluster):
    cluster.append(point)

    for neighbor in neighbors:
        if neighbor not in visited:
            visited.add(neighbor)
            new_neighbors = region_query(neighbor)

            if len(new_neighbors) >= minPts:
                neighbors.extend(new_neighbors)

        if neighbor not in cluster:
            cluster.append(neighbor)

for point in data:
    if point in visited:
        continue

    visited.add(point)
    neighbors = region_query(point)

    if len(neighbors) < minPts:
        noise.append(point)
    else:
        cluster = []
        expand_cluster(point, neighbors, cluster)
        clusters.append(cluster)

print("Clusters :", clusters)
print("Noise :", noise)
```
