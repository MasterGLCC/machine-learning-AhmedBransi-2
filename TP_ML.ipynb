```python
# =========================================
# TP MACHINE LEARNING - ETUDE DES REGRESSIONS
# Objectif : Prédire la note des étudiants
# =========================================

# ---------------------------
# 1. Importation des bibliothèques
# ---------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 2. Charger le dataset
# ---------------------------
data = pd.read_csv("dataset_etudiants.csv")

print("Aperçu du dataset :")
print(data.head())

# ---------------------------
```

```python
# 3. Régression linéaire simple
# ---------------------------
# Note = f(Heures_etude)

X_simple = data[['Heures_etude']]
y = data['Note']

model_simple = LinearRegression()
model_simple.fit(X_simple, y)

y_pred_simple = model_simple.predict(X_simple)
```

```python
# Graphique
plt.scatter(X_simple, y)
plt.plot(X_simple, y_pred_simple)
plt.title("Régression Linéaire Simple")
plt.xlabel("Heures d'étude")
plt.ylabel("Note")
plt.show()

print("\n--- Régression Linéaire Simple ---")
print("Coefficient:", model_simple.coef_[0])
print("Intercept:", model_simple.intercept_)
print("R2:", r2_score(y, y_pred_simple))
print("MSE:", mean_squared_error(y, y_pred_simple))

# ---------------------------
```

```python
# 4. Régression linéaire multiple
# ---------------------------
# Note = f(Heures_etude, Heures_sommeil, Participation)

X_multi = data[['Heures_etude', 'Heures_sommeil', 'Participation']]

model_multi = LinearRegression()
model_multi.fit(X_multi, y)

y_pred_multi = model_multi.predict(X_multi)

print("\n--- Régression Linéaire Multiple ---")
print("Coefficients:", model_multi.coef_)
print("Intercept:", model_multi.intercept_)
print("R2:", r2_score(y, y_pred_multi))
print("MSE:", mean_squared_error(y, y_pred_multi))

# ---------------------------
```

```python
# 5. Régression polynomiale
# ---------------------------
# Note = f(Heures_etude^2)

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_simple)

model_poly = LinearRegression()
model_poly.fit(X_poly, y)

y_pred_poly = model_poly.predict(X_poly)
```

```python
# Graphique
plt.scatter(X_simple, y)
plt.plot(X_simple, y_pred_poly)
plt.title("Régression Polynomiale")
plt.xlabel("Heures d'étude")
plt.ylabel("Note")
plt.show()

print("\n--- Régression Polynomiale ---")
print("Coefficients:", model_poly.coef_)
print("Intercept:", model_poly.intercept_)
print("R2:", r2_score(y, y_pred_poly))
print("MSE:", mean_squared_error(y, y_pred_poly))
```
