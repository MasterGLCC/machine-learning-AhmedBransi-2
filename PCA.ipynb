```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 1. DONNÉES
# ==========================================================
# Taille / Poids
X = np.array([
    [170, 65],
    [180, 80],
    [175, 75],
    [160, 50],
    [165, 55],
    [185, 90]
])

# ==========================================================
# 2. CENTRAGE DES DONNÉES
# ==========================================================
# On soustrait la moyenne
# Important pour PCA

mean = np.mean(X, axis=0)

X_centered = X - mean

print("Moyenne :")
print(mean)

print("\nDonnées centrées :")
print(X_centered)


# ==========================================================
# 3. MATRICE DE COVARIANCE
# ==========================================================
# Elle mesure la relation entre variables

cov_matrix = np.cov(X_centered.T)

print("\nMatrice covariance :")
print(cov_matrix)


# ==========================================================
# 4. VALEURS PROPRES / VECTEURS PROPRES
# ==========================================================
# Eigenvalues  = importance
# Eigenvectors = directions principales

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\nValeurs propres :")
print(eigenvalues)

print("\nVecteurs propres :")
print(eigenvectors)


# ==========================================================
# 5. TRI DES COMPOSANTES
# ==========================================================
# On trie du plus important au moins important

sorted_index = np.argsort(eigenvalues)[::-1]

eigenvalues = eigenvalues[sorted_index]
eigenvectors = eigenvectors[:, sorted_index]

print("\nValeurs propres triées :")
print(eigenvalues)


# ==========================================================
# 6. SÉLECTION DES COMPOSANTES PRINCIPALES
# ==========================================================
# Ici on garde seulement 1 composante principale

k = 1

principal_components = eigenvectors[:, :k]

print("\nComposantes principales :")
print(principal_components)


# ==========================================================
# 7. PROJECTION DES DONNÉES
# ==========================================================
# Transformation des données dans le nouvel espace

X_pca = np.dot(X_centered, principal_components)

print("\nDonnées après PCA :")
print(X_pca)


# ==========================================================
# 8. VISUALISATION
# ==========================================================
plt.scatter(X[:,0], X[:,1])

plt.xlabel("Taille")
plt.ylabel("Poids")
plt.title("Données originales")

plt.grid()
plt.show()
```
